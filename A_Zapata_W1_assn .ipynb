## Urban Index per State



#### *Introduction to the dataset* 

The data set was obtained from fivethirtyeight.com website. On this website the dataset is used to measure the urban index obtained from the data. They used the urban index to analyze the relationship of urban and rural ideas and their political incline towards democratic or republican affiliation.

#### *Explanation of datase*

The dataset includes two .csv files. One with only the state and the urban index obatined from the natural logarithm of the average number of people living within a five-mile radius of a given resident in each census tract in that state.

The second .csv file has more information on the census tracts in the us, their location, population, distance of residents to census tracts and the calculaton of urban index per census tracts.
